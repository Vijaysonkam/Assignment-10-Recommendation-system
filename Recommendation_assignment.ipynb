```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('book_1.csv', encoding='unicode_escape')
```

```python
df
```

```python
df=df.drop(['name'], axis=1)
```

```python
len(df.userid.unique())
```

```python
len(df.booktitle.unique())
```

```python
books_df= df.pivot_table(index='userid',columns='booktitle', values='bookrating').reset_index(drop=True)
```

```python
books_df
```

```python
books_df.index=df.userid.unique()
```

```python
books_df
```

```python
# imput thos NaN values with 0
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(books_df.values, metric='cosine')
```

```python
user_sim
```

```python
# store the result in dataframe
user_sim_df= pd.DataFrame(user_sim)
```

```python
user_sim_df.index=df.userid.unique()
user_sim_df.columns=df.userid.unique()
```

```python
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
